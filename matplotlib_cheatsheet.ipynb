### Basic

```python
x1 = np.linspace(0, 5, 10)
y1 = x1 ** 2
plt.plot(x1, y1) # quick plotting
plt.title('This is a title') # add a title
plt.xlabel('Days') # add x-label
plt.ylabel('Days Squared') # add y-label
plt.show() # when not in jupyter-notebook
```

### Multiple plots

```python
plt.subplot(1, 2, 1) # 1 row; 2 plots; 1st plot
plt.plot(x1, y1, 'r') # 'r' for red

plt.subplot(1, 2, 2) # 1 row; 2 plots; 2nd plot
plt.plot(x1, y1, 'b') # 'b' for blue
```

### Figure objects (Figure object generated - require sub-plots to be added)

```python
figObj = plt.figure(figsize = (5, 4), dpi = 100) # 'figsize' is width and height in inches; 'dpi' is dots per inch
axes1 = figObj.add_axes([0.1, 0.1, 0.9, 0.9]) # rect indicates [x-coord of axes, y-coord of axes, width of axes, height of axes] wrt to the figure object
axes1.set_xlabel('Days') # set x-label
axes1.set_ylabel('Days Squared') # set y-label
axes1.set_title('THis is a title') # set title
axes1.plot(x1, y1, label = 'x/x ** 2') # plot a graph
axes1.plot(y1, x1, label = 'x **2/x') # plot another graph
axes1.legend(loc = 0) # 0: auto select best location for legend

# add more axes
axes2 = figObj.add_axes([0.45, 0.45, 0.4, 0.4])
axes2.set_xlabel('Days') # set x-label
axes2.set_ylabel('Days Squared') # set y-label
axes2.set_title('This is a title') # set title
axes2.plot(x1, y1, label = 'x/x ** 2') # plot a graph
axes2.plot(y1, x1, label = 'x **2/x') # plot another graph

axes2.text(0.5, 0.5, 'Message') # add a text wrt to co-ordinates of the figure
```

### Subplots (Figure object and subplots can be generated)

```python
figObj2, axes2 = plt.subplots(figsize = (8, 8), nrows = 1, ncols = 3) 
# there will be 1 row and 3 columns. axes2 will be an array of subplots
plt.tight_layout() # automatically add padding between subplots
axes2[1].set_title('Subplot 2') # add title to 2nd subplot
axes2[1].set_xlabel('X') #add xlabel to the 2nd plot
axes2[1].set_ylabel('X Squared') # add y-label to the 2nd plot
axes2[1].plot(x1, y1) # add graph to the 2nd plot
```

### Appearance options

```python
# default colors (b: blue; g: green; r: red; c: cyan; m: magenta; y: yellow; k: black; w: white)
# color = 0.75 creates a 75% of that color
# You can use hexcodes color = '#eeefff'
figObj3 = plt.figure(figsize = (6, 4))
axes3 = figObj3.add_axes([0, 0, 1, 1])
axes3.plot(x1, y1, color = 'navy', alpha = 1, lw = 2, ls = '-.', marker = 'x', markersize = 7, 
          markerfacecolor = 'y', markeredgecolor = 'green', markeredgewidth = 4)
# alpha is transparency; lw is line width; ls is line style
axes3.set_xlim([0, 3]) # set the limit on x axis
axes3.set_ylim([0, 8]) # set the limit on y axis
axes3.grid(True, color = '0.6', dashes = (7, 2, 15, 2)) # set the grid background -> 7 points 2 spaces 15 points 2 spaces
axes3.set_facecolor('#FAEBD7') # set background color
```

### Pie Charts

```python
# A pie-chart has categories and each category has a count. 
# Each category is represented as a wedge and each wedge can be color coded
import random
categories = ['Water', 'Normal', 'Flying', 'Grass', 'Psychic', 'Bug', 'Fire', 'Poison',
'Ground', 'Rock', 'Fighting', 'Dark', 'Steel', 'Electric', 'Dragon', 'Fairy',
'Ghost', 'Ice']
count = [133, 109, 101, 98, 85, 77, 68, 66, 65, 60, 57, 54, 53, 51, 50, 50, 46, 40]
wedgeColors = []
for i in range(len(categories)):
    wedgeColors.append((random.uniform(0, .5), random.uniform(0, .5), random.uniform(0, .5)))

# we can pull out the wedges using explode attribute
explode = [0] * len(categories)
# exploding the first 2 wedges
explode[0] = 0.3
explode[1] = 0.3

wedges, texts, autotexts = plt.pie(count, labels = categories, explode = explode, colors = wedgeColors, 
                                  autopct = '%1.0f%%', shadow = True, startangle = 140, textprops = dict(color = 'w'))

# autopct to calculate a percentage out of counts attribute; textprops is a dict of options under which color is set to white
plt.legend(wedges, categories, loc = 'right', bbox_to_anchor = (1, 0, 0.5, 1))
# use bbox_to_anchor to place legend wrt to plot
```

### About matplotlib figures
1. matplotlib has something to do with state management. 'plt.figure()' sets creates a figure and sets the current figure to this figure
2. plt.plot (or any other plotting methods creates a figure object, sets the current figure object, creates an axes, sets the current axes and then plots the graph on the generated figure and axes objects)
3. As fas the jupyter notebook application is concerned, scope for plt.gca() (get current axes) plt.gcf() (get current figure) methods is the cell they are written. They will return the details if any figure or axes are set in that particular cell

### Tables

```python
# using the same time series dataset
timeData = pd.read_csv('timeseries.txt', delimiter = ',') # timeseries fetched from yahoo finance website
timeData['Open'] = pd.Series([round(val, 2) for val in timeData['Open']], index = timeData.index)
timeData['High'] = pd.Series([round(val, 2) for val in timeData['High']], index = timeData.index)
timeData['Low'] = pd.Series([round(val, 2) for val in timeData['Low']], index = timeData.index)
timeData['Close'] = pd.Series([round(val, 2) for val in timeData['Close']], index = timeData.index)
timeData['Adj Close'] = pd.Series([round(val, 2) for val in timeData['Adj Close']], index = timeData.index)

stockData = timeData[-5:]
columns = ('Date', 'Open', 'High', 'Low', 'Close', 'Adj Close', 'Volume')
stockData_np = stockData.to_numpy()

# creating a figure
figTable = plt.figure(linewidth = 2, tight_layout = {'pad': .5}, figsize = (8, 8))
# plt.table() creates an axes and then creates table inside the axes, then the table is not rendered in an user friendly way. Hence we need to disable the axes before plotting the table
axesTable = plt.gca() # get current axes. Creates one if there arent any
axesTable.get_xaxis().set_visible(False)
axesTable.get_yaxis().set_visible(False)
plt.box(on = None)
# set the color map = cmap
colors = plt.cm.Blues(np.full(len(columns), 0.2))
# create the table
tableFigure = plt.table(cellText = stockData_np, loc = 'center', colLabels = columns,
                        colColours = colors)
tableFigure.set_fontsize(18)
tableFigure.scale(3, 2.5)
```

### 3D Plots

```python
from mpl_toolkits import mplot3d
fig3D = plt.figure(figsize = (8, 5), dpi = 100)
axes3D = fig3D.add_axes([0.1, 0.1, 0.9, 0.9], projection = '3d')
zAxis = 15 * np.random.random(100)
xAxis = np.sin(zAxis) * np.random.random(100)
yAxis = np.sin(zAxis) * np.random.random(100)
# 3D scatter
#axes3D.scatter3D(xAxis, yAxis, zAxis, c = zAxis, cmap = 'Blues')


# contour - contour needs an axis that is a function of other 2 axis in a 3 plot
def getZ(x, y):
    return np.sin(np.sqrt(x**2 + y**2))
xContour = np.linspace(-6, 6, 30)
yContour = np.linspace(-6, 6, 30)
xContour, yContour = np.meshgrid(xContour, yContour) # returns all possible combinations of x & y values
zContour = getZ(xContour, yContour) # this will fetch the z value for every possible x & y combination returned from above
axes3D.view_init(55, 55) # set the viewing angle for the plot
#axes3D.contour3D(xContour, yContour, zContour, 100, cmap = 'Blues') # 100 represents the contour thickness

axes3D.set_xlabel('X')
axes3D.set_ylabel('Y')
axes3D.set_zlabel('Z')

axes3D.plot_wireframe(xContour, yContour, zContour, color = 'blue') # wireframe design
axes3D.plot_surface(xContour, yContour, zContour, rstride = 1, cstride = 2, cmap = 'Blues', edgecolor = 'none')
```

### Heatmaps

```python
# A heatmap is a color coded representation of data from a 2D list. Used for heirarchical clustering
symptoms = ["Coronavirus","Influenza","Pneumonia","Dyspnea"]
dates = ["Jun28","Jul05","Jul12","Jul19","Jul26","Aug02","Aug09","Aug16","Aug21"]
symp_per = np.array([[5.2, 5.5, 5.7, 5.6, 5.3, 5.1, 5.0, 4.9, 5.3],
                    [3.5, 4.0, 4.3, 3.9, 3.5, 3.2, 2.7, 2.2, 2.0],
                    [1.8, 2.2, 2.3, 2.2, 2.1, 1.9, 1.7, 1.4, 1.3],
                    [1.0, 1.1, 1.1, 1.0, 0.9, 0.8, 0.8, 0.8, 0.7]])

figHeatmap, axesHeatmap = plt.subplots() # default fig size and default # of subplots
im = axesHeatmap.imshow(symp_per, cmap = 'Wistia')
axesHeatmap.set_xticks(np.arange(len(dates))) # set # of x ticks
axesHeatmap.set_xticklabels(dates) # set labels for the ticks
axesHeatmap.set_yticks(np.arange(len(symptoms))) # set # of y ticks
axesHeatmap.set_yticklabels(symptoms) # set labels for the ticks
# Rotate labels on the bottom so they don't overlap
plt.setp(axesHeatmap.get_xticklabels(), rotation=45, ha="right",
         rotation_mode="anchor")

# add text annotations

for i in range(len(symptoms)):
    for j in range(len(dates)):
        text = axesHeatmap.text(j, i, symp_per[i, j], ha = 'center', va = 'center', color = 'k', fontweight = 'bold')
```
